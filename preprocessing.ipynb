```python
def preprocessing(path):
    df = pd.read_csv(path)
    df["WGS84 N"]=0
    df["WGS84 E"]=0
    df = df[df["Rok"]==2022]
    df = df[df["avg time"]=="1g"]
    df = df.reset_index(drop=True)
    return df
```

```python
def combine_geo(df, stations):
    for i in range(len(df)):
        station_name = df.loc[i,"Kod stacji"]
        station_loc = stations[stations["Kod stacji"]==station_name]
        df.loc[i, "WGS84 N"] = station_loc.iloc[0, 1]
        df.loc[i, "WGS84 E"] = station_loc.iloc[0, 2]
    return df
```

```python
pm10_2022_1h = preprocessing(path="data/PM10.csv")
pm10_2022_1h = combine_geo(df=pm10_2022_24h, stations=stations)
pm10_2022_1h.to_csv("data/PM10_geo.csv")
```
